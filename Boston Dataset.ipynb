```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
import operator
from sklearn.datasets import load_boston
boston_dataset = load_boston()
```

```python
boston = pd.DataFrame(boston_dataset.data, columns = boston_dataset.feature_names)
```

```python
boston
```

```python
boston['target']=boston_dataset.target 
```

```python
boston
```

```python
plt.scatter(boston.LSTAT, boston.target)
```

```python
x = boston['LSTAT']
y = boston['target']
```

```python
msk = np.random.rand(len(boston)) < 0.8
```

```python
len(msk)
```

```python
x_train = x[msk]
y_train = y[msk]
x_test = x[~msk]
y_test = y[~msk]
```

```python
x_train = x_train[:,np.newaxis]
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
y_pred = regr.predict(x_train)
```

```python
plt.plot(x_train, y_train, 'ro')
plt.plot(x_train, y_pred)
plt.show()
```

```python
poly = PolynomialFeatures(8)
```

```python
x_train_poly = poly.fit_transform(x_train)
```

```python
regr.fit(x_train_poly, y_train)
y_poly_predict = regr.predict(x_train_poly)
```

```python
plt.plot(x_train, y_train, 'ro')
plt.plot(x_train, y_poly_predict)
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(y_train,y_poly_predict))
r2 = r2_score(y_train,y_poly_predict)
print(rmse)
print(r2)
```

```python
x_test = x_test[:,np.newaxis]
y_test_predict = regr.predict(poly.fit_transform(x_test))
```
